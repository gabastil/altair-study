```python
import altair as alt
import pandas as pd
import geopandas as gpd
```

### Load Data

```python
data = pd.read_csv('../../odi-covid-19/data/datasets/demographics/demographics_age_gender.csv')
data.head()
```

### Define Charts

```python
bar_1 = alt.Chart(data).mark_bar()
bar_2 = alt.Chart(data).mark_bar().encode(x='ADM2_NAME:N')
```

---
### Plot and Sort Data

##### Sort Quantitative (Q) Data

```python
bar_1.encode(y=alt.Y('Population:Q'), 
             x=alt.X('ADM2_NAME:N', sort=alt.EncodingSortField('Population', op='sum', order='descending')))
```

##### Sort Nominative (N) Data

```python
bar_2.encode(x='Population:Q', y=alt.Y('ADM2_NAME:N', sort=alt.EncodingSortField('ADM2_NAME', order='descending')))
```

##### Sort Faceted Data - Individual Chart (N)

Sort individual axes

```python
bar_1.encode(x=alt.X('Gender:N', sort=alt.EncodingSortField('Gender', order='descending')), 
             y='Population:Q', column='Age',
            ).transform_filter(
    alt.FieldOneOfPredicate('Gender', oneOf=['Male', 'Female'])
).configure_facet(spacing=1).properties(width=10)
```

Sort faceted charts

```python
facets = bar_1.encode(alt.X('Gender:N', sort=alt.EncodingSortField('Gender', order='ascending')),
                      alt.Y('Population:Q'),
                      alt.Column('Age', sort=alt.EncodingSortField('Age', order='descending'))
                     ).transform_filter(alt.FieldOneOfPredicate('Gender', ['Male', 'Female']))

facets.configure_facet(spacing=1).properties(width=10)
```

```python
facets = bar_1.encode(alt.X('Gender:N', sort=alt.EncodingSortField('Population', order='descending')),
                      alt.Y('Population:Q'),
                      alt.Column('Age', sort=alt.EncodingSortField('Population', order='descending'))
                     ).transform_filter(alt.FieldOneOfPredicate('Gender', ['Male', 'Female']))

facets.configure_facet(spacing=1).properties(width=10)
```
